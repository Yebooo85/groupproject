```python
import json
file = '../json/project-10.json'
outputfile = '../json/grouped_data_10.json'

# 加载JSON数据
with open(file, 'r', encoding='utf-8') as f:
    data = json.load(f)
```

```python
# 用代码加标签的用这段代码

# 改进的分组函数，按metric-value-unit分组
def group_data(data):
    grouped_data = []
    current_entry = {"metric": None, "value": None, "unit": None}
    
    for entry in data:
        for item in entry.get("metrics", []):
            # 根据标签填充对应字段
            if "metric" in item["labels"]:
                # 如果当前entry有完整记录，则保存并重置
                if any(current_entry.values()):
                    grouped_data.append(current_entry)
                    current_entry = {"metric": None, "value": None, "unit": None}
                current_entry["metric"] = item["text"]
            elif "value" in item["labels"]:
                current_entry["value"] = item["text"]
            elif "unit" in item["labels"]:
                current_entry["unit"] = item["text"]
        
    # 将最后的条目添加到分组数据中（如果非空）
    if any(current_entry.values()):
        grouped_data.append(current_entry)

    return grouped_data
```

```python
# 用的自动加标签的用这段代码
def group_data(data):
    grouped_data = []
    current_entry = {"metric": None, "value": None, "unit": None}
    
    for entry in data:
        for item in entry.get("label", []):  # 确保使用正确的标签键
            label_type = item["labels"][0] if item["labels"] else None  # 获取第一个标签类型
            
            if label_type == "metric":
                if any(current_entry.values()):
                    grouped_data.append(current_entry)
                    current_entry = {"metric": None, "value": None, "unit": None}
                current_entry["metric"] = item["text"]
            elif label_type == "value":
                current_entry["value"] = item["text"]
            elif label_type == "unit":
                current_entry["unit"] = item["text"]

        # 确保当前条目在每个 entry 完成后添加
        if any(current_entry.values()):
            grouped_data.append(current_entry)
            current_entry = {"metric": None, "value": None, "unit": None}
    
    return grouped_data
```

```python
# 分组数据
grouped_data = group_data(data)

# 过滤掉没有value的条目
filtered_data = [entry for entry in grouped_data if entry["value"] is not None]

# 保存分组后的数据到新的JSON文件
with open(outputfile, 'w', encoding='utf-8') as f:
    json.dump(filtered_data, f, ensure_ascii=False, indent=4)

# 输出分组后的数据
print(json.dumps(filtered_data, ensure_ascii=False, indent=4))
```
